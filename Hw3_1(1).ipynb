Удалите из таблицы данные по всей стране, последняя строчка таблицы

```python
tabl_koef = tabl_koef.drop([27])
print(tabl_koef)
```

Замените отсутствующие данные в столбцах средними значениями по этим столбцам (метод fillna)

```python
tabl_koef = tabl_koef.fillna({"1950": tabl_koef["1950"].mean(), 
                  "1960": tabl_koef["1960"].mean(), 
                  "1970": tabl_koef["1970"].mean(), 
                  "2014": tabl_koef["2014"].mean(), 
                  "2019": tabl_koef["2019"].mean()})
print(tabl_koef)
```

```python
def averages(year):
  average = []
  for t in year:
    avg = t.mean()
    average.append(avg)
  return average 

year = [tabl_koef['1950'], tabl_koef['1960'], tabl_koef['1970'], tabl_koef['1990'], tabl_koef['2000'], tabl_koef['2012'], tabl_koef['2014'], tabl_koef['2019']]
data = averages(year)
```

```python
labels = ["1950","1960", "1970","1990","2000","2012", "2014", "2019"]

data = data
explode = [0]*len(labels)
plt.pie(
    data,
    shadow=True,
    explode=explode,
    autopct="%.2f%%",
    startangle=140,
    pctdistance=0.8,
    labeldistance=1.05,
)
fig = plt.gcf()
fig.set_size_inches(20,14)
plt.title('Percentage birth rate by year')
plt.legend( labels, title="Year", loc="upper left", bbox_to_anchor=(1, 0.1, 0.1, 1))
```

```python
Year = labels
data = data

plt.bar(Year, data, color = 'green')
plt.xlabel('YEAR', fontsize=14)
plt.ylabel('Average birth rate in all districts', fontsize=14)
plt.title('Average birth rate by year', fontsize=14)
fig = plt.gcf()
fig.set_size_inches(10,7)
plt.show()
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize=(20, 14), subplot_kw=dict(aspect="equal"), dpi= 80)

labels = tabl_koef['Регіон']
data = tabl_koef['1990']
explode = [0.15] + [0]*26

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}% ({:d} )".format(pct, absolute)

wedges, texts, autotexts = ax.pie(data, 
                                  autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="black"), 
                                  colors=plt.cm.Dark2.colors,
                                  startangle=140,
                                  explode=explode,
                                  pctdistance=1.05,
                                  labeldistance=1.35,
                                  shadow=True,)

ax.legend(wedges, labels, title="Region", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=10, weight=700)
ax.set_title("Birth rate by region in 1990")
```

Получите список регионов, где уровень рождаемости в 2019 году был выше среднего по Украине

```python
new_tab = pd.concat([tabl_koef["Регіон"], tabl_koef["2019"]], axis = 1)

new_tab = new_tab[new_tab["2019"] > new_tab["2019"].mean()]
reg_list = []
for reg in new_tab['Регіон']:
  reg_list.append(reg)
print(reg_list)
```

В каком регионе была самая высокая рождаемость в 2014 году?

```python
new_tab = pd.concat([tabl_koef["Регіон"], tabl_koef["2014"]], axis = 1)

new_tab[new_tab["2014"] >=  new_tab['2014'].max()]
```
