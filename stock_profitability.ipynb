В таблице приведены данные о взаимонезависимых доходностях трех акций: A, B, C.

image.png

Постройте линии регрессии, с помощью функции regplot, зависимости доходности одной акции от другой. Всего три графика. Выберите пару акций которые наименее зависимы в доходностях от друг друга.

```python
import pandas as pd
import seaborn as sns
```

```python
moment = [1, 2, 3, 4, 5, 6]
A = [25, -10, 10, 5, 35, 13]
B = [0, 15, -5, 5, 20, 25]
C = [10, 25, -15, -5, -5, 15]
data = pd.DataFrame({'moment':moment, 'A': A, 'B': B, 'C': C})
data
```

```python
sns.regplot(x = "A",
            y = "B",
            data = data)
```

```python
sns.regplot(x = "A",
            y = "C",
            data = data, 
            scatter_kws={"color": "black"}, 
            line_kws={"color": "red"})
```

```python
sns.regplot(x = "C",
            y = "B",
            data = data, 
            color = 'green')
```
